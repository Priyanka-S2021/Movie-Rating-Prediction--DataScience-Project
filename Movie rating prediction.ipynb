```python
import pandas as pd

# Strip any extra spaces from the column names
df.columns = df.columns.str.strip()

# Ensure 'Year' is numeric and handle errors
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Calculate movie age (2025 - Year)
df['movie_age'] = 2025 - df['Year']

# Determine if a movie is old (before the year 2000)
df['is_old_movie'] = df['Year'].apply(lambda x: 1 if pd.notna(x) and x < 2000 else 0)

# Ensure 'Rating' and 'Votes' are numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

# Genre Popularity (average rating by genre)
# Ensure 'Genre' is treated as a categorical type but 'Rating' is numeric for aggregation
genre_avg_rating = df.groupby('Genre')['Rating'].mean().to_dict()
df['genre_avg_rating'] = df['Genre'].map(genre_avg_rating)

# Director Success Rate (average rating by director)
director_avg_rating = df.groupby('Director')['Rating'].mean().to_dict()
df['director_success_rate'] = df['Director'].map(director_avg_rating)

# Lead Actor Popularity (average votes by actor)
# 'Actor 1' can be treated as a string, but ensure 'Votes' is numeric for aggregation
actor1_avg_votes = df.groupby('Actor 1')['Votes'].mean().to_dict()
df['actor1_popularity'] = df['Actor 1'].map(actor1_avg_votes)

# Actor Combination Popularity (count of same actor combinations)
# Ensure proper handling of NaN values and type conversion to string
df['actor_combo'] = (
    df['Actor 1'].fillna('').astype(str) + '_' +
    df['Actor 2'].fillna('').astype(str) + '_' +
    df['Actor 3'].fillna('').astype(str)
)

# Count combinations
combo_counts = df['actor_combo'].value_counts().to_dict()
df['actor_combo_popularity'] = df['actor_combo'].map(combo_counts)

# Missing Director or Actor Flags
df['missing_director'] = df['Director'].isnull().astype(int)
df['missing_actor1'] = df['Actor 1'].isnull().astype(int)
df['missing_actor2'] = df['Actor 2'].isnull().astype(int)
df['missing_actor3'] = df['Actor 3'].isnull().astype(int)

# --- Final Cleaning ---
# Fill any NaNs created during mapping with median values
for col in ['director_success_rate', 'actor1_popularity', 'actor_combo_popularity']:
    df[col].fillna(df[col].median())

# Drop the original categorical columns that are not needed (optional)
df.drop(['Name', 'Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3', 'actor_combo'], axis=1, inplace=True)
```

```python
#Handle Missing Values

from sklearn.impute import SimpleImputer

# Assuming df is your DataFrame
numerical_cols = ['Rating', 'Votes', 'genre_avg_rating', 'director_success_rate', 'actor1_popularity', 'actor_combo_popularity']
categorical_cols = ['Year', 'is_old_movie', 'missing_director', 'missing_actor1', 'missing_actor2', 'missing_actor3']

# For numerical features, replace missing with the mean
numerical_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = numerical_imputer.fit_transform(df[numerical_cols])

# For categorical features, replace missing with the mode
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
```

```python
#Encoding Categorical Variables

from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# For binary categorical variables (like is_old_movie), use LabelEncoder
label_encoder = LabelEncoder()
df['is_old_movie'] = label_encoder.fit_transform(df['is_old_movie'])

```

```python
#Scaling Numerical Variables

from sklearn.preprocessing import StandardScaler

# List of numerical columns you want to scale
numerical_cols = ['Rating', 'Votes', 'movie_age', 'genre_avg_rating', 'director_success_rate', 
                  'actor1_popularity', 'actor_combo_popularity']

# Initialize the scaler
scaler = StandardScaler()

# Scale the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
#MODEL SELECTION

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

X = df.drop('Rating', axis=1)
y = df['Rating']


# Step 1: Impute missing values
imputer = SimpleImputer(strategy='mean')  # or use 'median' if skewed
X_imputed = imputer.fit_transform(X)

# Step 2: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Step 3: Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


```

```python
#MODEL SELECTION -BEST MODEL COMPARE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'XGBoost': XGBRegressor(random_state=42)
}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    
    print(f"\n{name} Performance:")
    print(f"MAE: {mean_absolute_error(y_test, y_pred):.3f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.3f}")
    print(f"R Score: {r2_score(y_test, y_pred):.3f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

# Corrected parameter grid
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Initialize model
rf = RandomForestRegressor(random_state=42)

# Perform Randomized Search
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=10,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1,
    scoring='neg_mean_squared_error'
)

random_search.fit(X_train_scaled, y_train)
best_rf = random_search.best_estimator_
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = best_rf.predict(X_test_scaled)

print("Tuned Random Forest Performance:")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.3f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.3f}")
print(f"R Score: {r2_score(y_test, y_pred):.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract feature importances
features = X.columns
importances = best_rf.feature_importances_

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features)
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.tight_layout()
plt.show()

import shap

# Create a SHAP explainer object for the Random Forest model
explainer = shap.TreeExplainer(best_rf)

# Compute SHAP values for the test set
shap_values = explainer.shap_values(X_test_scaled)

# Generate a SHAP summary plot to visualize feature importance
shap.summary_plot(shap_values, X_test_scaled, feature_names=features)

```
